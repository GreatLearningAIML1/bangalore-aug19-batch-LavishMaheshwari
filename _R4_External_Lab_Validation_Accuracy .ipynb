### Q2.Prepare the model data by converting non-numeric to dummy ( 1 Marks)
##### Hint: Use get_dummies

```python
# Print Shape of modedl data
df_dummies = pd.get_dummies(df, columns = ['default', 'account_check_status', 'duration_in_month',
       'credit_history', 'purpose', 'credit_amount', 'savings',
       'present_emp_since', 'installment_as_income_perc',
       'personal_status_sex', 'other_debtors', 'present_res_since', 'property',
       'age', 'other_installment_plans', 'housing', 'credits_this_bank', 'job',
       'people_under_maintenance', 'telephone', 'foreign_worker'], drop_first = True)
df_dummies.shape
```

### Q3 Split Train/Test data 70:30 ratio( 1 Marks)
##### Hint:from sklearn.model_selection import train_test_split

```python
from sklearn import model_selection
test_size = 0.30 # taking 70:30 training and test set
X=df_dummies.drop('default_1',axis=1)
y=df_dummies['default_1']
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=test_size)
```

### Q4 Build Random Forest Model( 1 Marks)
#### Hint:from sklearn.ensemble import RandomForestClassifier using n_jobs=2,n_estimators=500,criterion="entropy",random_state=9999

```python
from sklearn.ensemble import  RandomForestClassifier 
clf = RandomForestClassifier(n_jobs=2, n_estimators=500,criterion="entropy", random_state=9999)
clf.fit(X_train, y_train)
```

### Q5 Calculate Confusion Matrix and Accuracy score (1 Marks)
##### Hint: Use confusion_matrix and accuracy_score

```python
from sklearn.metrics import confusion_matrix 
from sklearn.metrics import accuracy_score 
predicted= clf.predict(X_test)
print("Confusion Matrix: ", confusion_matrix(y_test, predicted)) 
print ("Accuracy : ", accuracy_score(y_test,predicted)*100) 
```
